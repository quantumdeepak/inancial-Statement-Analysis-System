```python
# Import required libraries
from llama_index.llms.gemini import Gemini
from llama_index.core import Settings, StorageContext, VectorStoreIndex
from llama_index.core.node_parser import SimpleNodeParser  # Changed from MarkdownElementNodeParser
from llama_index.embeddings.gemini import GeminiEmbedding
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_parse import LlamaParse
import chromadb
import nest_asyncio
import os
from typing import List
import logging
from getpass import getpass
import google.generativeai as genai

# Enable async operations in Jupyter notebooks
nest_asyncio.apply()

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

class DocumentProcessor:
    def __init__(self, gemini_api_key: str, llama_cloud_api_key: str):
        """
        Initialize the document processor with necessary configurations
        """
        self.gemini_api_key = gemini_api_key
        self.llama_cloud_api_key = llama_cloud_api_key

        # Configure Gemini
        genai.configure(api_key=self.gemini_api_key)

        # Initialize Gemini LLM with specific parameters
        self.llm = Gemini(
            model="models/gemini-1.5-flash",
            api_key=self.gemini_api_key,
            temperature=0.3,
            top_p=0.85,
            max_tokens=4096
        )

        # Initialize Gemini Embeddings
        self.embed_model = GeminiEmbedding(
            model="models/embedding-001",
            api_key=self.gemini_api_key,
            dimension=768  # Specify dimension for Gemini embeddings
        )

        # Set up global configurations
        Settings.llm = self.llm
        Settings.embed_model = self.embed_model

        # Initialize Chroma client
        self.chroma_client = chromadb.PersistentClient(path="./chroma_db")

    def parse_document(self, file_path: str) -> List:
        """
        Parse PDF document using LlamaParse
        """
        try:
            # Updated parsing with content guideline instead of deprecated instruction
            parser = LlamaParse(
                api_key=self.llama_cloud_api_key,
                result_type="markdown",
                content_guideline_instruction="Extract and preserve all text content, tables, and structural elements from the academic paper."
            )
            documents = parser.load_data(file_path)
            logger.info("Document parsed successfully")
            return documents
        except Exception as e:
            logger.error(f"Error parsing document: {e}")
            raise

    def process_nodes(self, documents: List) -> List:
        """
        Process documents into nodes using SimpleNodeParser
        """
        try:
            # Use SimpleNodeParser instead of MarkdownElementNodeParser
            node_parser = SimpleNodeParser.from_defaults(
                chunk_size=1024,
                chunk_overlap=200
            )

            # Get nodes from documents
            nodes = node_parser.get_nodes_from_documents(documents)

            logger.info(f"Processed {len(nodes)} nodes")
            return nodes

        except Exception as e:
            logger.error(f"Error processing nodes: {e}")
            raise

    def create_vector_store(self, collection_name: str):
        """
        Create a Chroma vector store
        """
        try:
            # Try to get existing collection or create new one
            try:
                collection = self.chroma_client.get_collection(name=collection_name)
            except:
                collection = self.chroma_client.create_collection(
                    name=collection_name,
                    metadata={"hnsw:space": "cosine"}  # Specify distance metric
                )

            vector_store = ChromaVectorStore(chroma_collection=collection)
            return vector_store
        except Exception as e:
            logger.error(f"Error creating vector store: {e}")
            raise

    def create_rag_pipeline(self, nodes: List, collection_name: str):
        """
        Create RAG pipeline with vector store
        """
        try:
            # Create vector store and storage context
            vector_store = self.create_vector_store(collection_name)
            storage_context = StorageContext.from_defaults(vector_store=vector_store)

            # Create index with all nodes
            index = VectorStoreIndex(
                nodes=nodes,
                storage_context=storage_context,
                show_progress=True  # Show progress during indexing
            )

            # Create query engine with specific parameters
            query_engine = index.as_query_engine(
                similarity_top_k=3,  # Number of similar chunks to retrieve
                response_mode="compact"  # More concise responses
            )

            logger.info("RAG pipeline created successfully")
            return query_engine
        except Exception as e:
            logger.error(f"Error creating RAG pipeline: {e}")
            raise

class QueryProcessor:
    def __init__(self, query_engine):
        self.query_engine = query_engine

    def process_query(self, query: str) -> str:
        """
        Process a query using the RAG pipeline
        """
        try:
            response = self.query_engine.query(query)
            return str(response)
        except Exception as e:
            logger.error(f"Error processing query: {e}")
            raise

def main():
    try:
        # Get API keys
        gemini_api_key = key
        llama_cloud_api_key = llama_key

        # Initialize processor
        processor = DocumentProcessor(
            gemini_api_key=gemini_api_key,
            llama_cloud_api_key=llama_cloud_api_key
        )

        # Parse document
        pdf_path = "/content/Sample Financial Statement.pdf"
        documents = processor.parse_document(pdf_path)

        # Process nodes
        nodes = processor.process_nodes(documents)

        # Create RAG pipeline
        query_engine = processor.create_rag_pipeline(nodes, "llm_recall_collection")

        # Initialize query processor
        query_processor = QueryProcessor(query_engine)

        # Example queries
        questions = [
            "Describe document",
            "What are the main findings about prompt dependency?",
            "Summarize the methodology used in the study"
        ]

        for question in questions:
            print(f"\nQuestion: {question}")
            response = query_processor.process_query(question)
            print(f"Answer: {response}\n")

    except Exception as e:
        logger.error(f"Error in main execution: {e}")
        raise

if __name__ == "__main__":
    main()
```
